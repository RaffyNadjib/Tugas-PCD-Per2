```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

gray = cv2.imread("bg.png", 0)

def quantize(img, bits):
    levels = 2**bits
    return np.floor(img / (256/levels)) * (256/levels)

q2 = quantize(gray, 2)
q4 = quantize(gray, 4)
q8 = quantize(gray, 8)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(q2, cmap='gray')
plt.title("2 bit (4 level)")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(q4, cmap='gray')
plt.title("4 bit (16 level)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(q8, cmap='gray')
plt.title("8 bit (256 level)")
plt.axis("off")

plt.show()

print("Jumlah nilai unik 8 bit:", len(np.unique(q8)))
print("Jumlah nilai unik 4 bit:", len(np.unique(q4)))
print("Jumlah nilai unik 2 bit:", len(np.unique(q2)))
```
